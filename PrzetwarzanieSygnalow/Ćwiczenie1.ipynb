# Zadanie 1.3
Sporządź wykres funkcji $$\begin{align*} 
y = x^2 + 5
\end{align*}$$ w dziedzinie od -10 do 10.

```python
from pylab import *
x = np.linspace(-10, 10)
y = x **2 + 5
figure()
plot(x, y, 'r')
xlabel('x')
h = ylabel('y')
h.set_rotation(0)
title(r'Wykres funkcji $y=x^2+5$')
show()
```

# Zadanie 1.4
Na jednym rysunku sporządź wykres funkcji sin(x) oraz cos(x), w dziedzinie od 0 do 2π.


```python
from pylab import *
import numpy as np
r=np.arange(0,2*np.pi,0.1)
y2 = np.sin(r)
y3 = np.cos(r)
sin = plot(r,y2,label='sin')
cos = plot(r,y3,label='cos')
legend()
xticks([0, pi, 2*pi],('0',r'$\pi$',r'$2\pi$'))
title('Wykres sinusa i cosinusa w przedziale od 0 do 2$\pi$')
show()
```

# Zadanie 1.5
Sporządź wykres sygnału sinusoidalnego o częstotliwości 1kHz i amplitudzie 2V w zakresie od 0 do 5ms.

```python
from pylab import *
import numpy as np
t=np.arange(0,0.005,0.00001) #time range
f=1000 #frequency in kHz
y = 2*np.sin(2*np.pi*f*t) #sine wave
xticks([0, 0.001, 0.002, 0.003, 0.004, 0.005])
xlabel('Czas[s]')
ylabel("Napięcie[V]")
title('Wykres sygnału sinusoidalnego o f=1kHz, amplitudzie 2V w zakresie 0-5ms')
plot(t,y)
show()
```
