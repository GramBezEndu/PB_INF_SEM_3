```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
x=np.linspace(-1,1,200)
y=np.sin(5*np.pi*x)
z=signal.sawtooth(5*np.pi*x)
c=signal.square(5*np.pi*x)
plt.stem(x,y)
plt.show()
plt.title("Sinus")
plt.xlabel("Argumenty")
plt.ylabel("Wartosci")
plt.stem(x,z)
plt.show()
plt.stem(x,c)
plt.show()
```
